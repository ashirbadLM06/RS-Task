3. Create an array from 1 to 12 and reshape it into:
- (3,4)
- (2,6)

7. Difference between Python list and NumPy array.

Python list:
- It contains different datatypes(Heterogeneous).
- slower execution speed.
- Require manual loops or list comprehensions.

Numpy array:
- It contains all elements of same types.(Homogeneous)
- Faster execution speed.
- packed contiguously in a single memory block.
- Supports direct element-wise vector operations.

9. What is the difference between:- dtype & astype()

dtype:- 

- check/return the data type of the object in numpy array series.

- It is an attribute.(No parenthesis required)

- It uses Zero Memory.

astype:-
- cast/convert to a new data type.

- It is a method.(Parenthesis required)
         
- It creates a new copy of your data in memory with the new type              

10. Create a 2D array and print:- shape- size- ndim- dtype
Change its datatype.

```python
a= np.arange(1,11).reshape(2,5)
print(a)
print("\n Shape:",a.shape)
print("\n Size:",a.size)
print("\n Dimension:",a.ndim)
print("\n Data types of the element:",a.dtype)
b=a.astype(float)
print("\n The change after changing of the data type:",b)
```
